```python
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.metrics import r2_score
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import accuracy_score
#X = data.drop(['Rating','Reviewer_Location','Review_Text','Year_Month_2010','Year_Month_2011','Year_Month_2012','Year_Month_2013','Year_Month_2014','Year_Month_2015','Year_Month_2016','Year_Month_2017','Year_Month_2018','Year_Month_2019','Month_1','Month_2','Month_3','Month_4','Month_5','Month_6','Month_7','Month_8','Month_9','Month_10','Month_11','Month_12'], axis = 1)
X = data.drop(['Rating','Reviewer_Location','Review_Text','Month_1','Month_2','Month_3','Month_4','Month_5','Month_6','Month_7','Month_8','Month_9','Month_10','Month_11','Month_12'], axis = 1)
#X = data.drop(['Rating','Reviewer_Location','Review_Text','Year_Month_2010','Year_Month_2011','Year_Month_2012','Year_Month_2013','Year_Month_2014','Year_Month_2015','Year_Month_2016','Year_Month_2017','Year_Month_2018','Year_Month_2019'], axis = 1)
#X = data.drop(['Rating','Reviewer_Location','Review_Text'], axis = 1)
Y = data['Rating']
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X) #Converts the dataframe into np.array of degree 3 polynomial and assigns to X_norm_poly

regr = LinearRegression()
regr.fit(X_poly, Y)
y_pred = regr.predict(X_poly)

print('Model Score:', metrics.r2_score(Y, y_pred)) 
```

```python
X.shape
```

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(X_poly, Y)
predicted_gaussian_y = nb.predict(X_poly)
print("Naive Bayes score 1: ",nb.score(X_poly, Y))
print("Accuracy Score:",accuracy_score(Y,predicted_gaussian_y))
```

```python
# KNN (K-Nearest Neighbors)
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier

knn_object = KNeighborsClassifier(n_neighbors = 5)     
knn_object.fit(X_poly, Y)
predicted_labels = knn_object.predict(X_poly)  

accuracy_score(Y, predicted_labels)
```

```python
# SVM (Support Vector Machine)

from sklearn import svm

svm_object = svm.SVC(kernel='linear') 
svm_object.fit(X_poly, Y)
predicted_labels = svm_object.predict(X_poly)  

accuracy_score(Y, predicted_labels)
```

```python
# Decision Tree

from sklearn.tree import DecisionTreeClassifier

dt_object=DecisionTreeClassifier(criterion='gini')
dt_object.fit(X, Y)
predicted_labels = dt_object.predict(X)  

accuracy_score(Y, predicted_labels)
```

```python
# Random Forest

from sklearn.ensemble import RandomForestClassifier

rf_object=RandomForestClassifier(n_estimators=50)
rf_object.fit(X, Y)
predicted_labels = rf_object.predict(X)  

accuracy_score(Y, predicted_labels)
```

```python
# xgboost

from xgboost import XGBClassifier

xgbc_object = XGBClassifier()
xgbc_object.fit(X_poly, Y)
predicted_labels = xgbc_object.predict(X_poly) 

accuracy_score(Y, predicted_labels)
```
